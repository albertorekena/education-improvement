```python
import numpy as np
import pandas as pd

from collections import Counter

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split

from sklearn.ensemble import IsolationForest

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report
from sklearn.metrics import mean_absolute_error,mean_squared_error

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',1000)
```

```python
df_por = pd.read_pickle('df_por-modelling-range_G3_5-repaired_G3_0_values.pkl')

df_por.head()
```

```python
df_por.shape
```

## Isolation Forest

```python
X = np.array(df_por.drop('G3',axis = 1))
```

```python
model = IsolationForest(random_state = 42)
model.fit(X)
```

```python
y_if = model.predict(X)
```

```python
y_if = pd.DataFrame(y_if)

y_if.columns = ['outlier']
```

```python
df_por = pd.concat([df_por,y_if],axis = 1)

df_por.head()
```

```python
print(df_por['outlier'].value_counts())
```

```python
print(df_por['outlier'].value_counts(normalize = True))
```

### Linear Regression

```python
X = np.array(df_por.drop('G3',axis = 1))
y = np.array(df_por['G3'])
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,stratify = y,random_state = 42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_test = pd.DataFrame(y_test)
y_pred = pd.DataFrame(y_pred)

df_temp = pd.concat([y_pred,y_test],axis = 1)
df_temp.columns = ['y_pred','y_test']

df_temp.head()
```

```python
plt.scatter(df_temp['y_test'],df_temp['y_pred'])
plt.xlabel('y_test',fontsize = 14)
plt.ylabel('y_pred',fontsize = 14)
plt.xlim(0,6)
plt.ylim(0,6)
plt.xticks(list(range(0,6)))
plt.yticks(list(range(0,6)))
plt.grid()
y_lim = plt.ylim()
x_lim = plt.xlim()
plt.plot(x_lim,y_lim,color = 'g',linewidth = 0.5);
```

```python
mean_absolute_error(y_pred,y_test)
```

```python
def rmse(y_true,y_pred):
    return np.sqrt(mean_squared_error(y_true,y_pred))
```

```python
rmse(y_pred,y_test)
```

### Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
y_test = pd.DataFrame(y_test)
y_pred = pd.DataFrame(y_pred)

df_temp = pd.concat([y_pred,y_test],axis = 1)
df_temp.columns = ['y_pred','y_test']

df_temp.head()
```

```python
c = Counter(zip(df_temp['y_pred'].tolist(),df_temp['y_test'].tolist()))
s = [15 * c[(x,y)] for x,y in zip(df_temp['y_pred'].tolist(),df_temp['y_test'].tolist())]

plt.scatter(df_temp['y_test'],df_temp['y_pred'],s = s)
plt.xlabel('y_test',fontsize = 14)
plt.ylabel('y_pred',fontsize = 14)
plt.xlim(0,6)
plt.ylim(0,6)
plt.xticks(list(range(0,6)))
plt.yticks(list(range(0,6)))
plt.grid()
y_lim = plt.ylim()
x_lim = plt.xlim()
plt.plot(x_lim,y_lim,color = 'g',linewidth = 0.5);
```

```python
mean_absolute_error(y_pred,y_test)
```

```python
rmse(y_pred,y_test)
```
